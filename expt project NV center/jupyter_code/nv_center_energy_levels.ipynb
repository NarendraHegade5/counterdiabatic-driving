```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
beta=1.0
y1= beta
def y2(alpha, beta):
    return 0.5* (beta- np.sqrt(beta**2+ 8*alpha**2))
def y3(alpha, beta):
    return 0.5* (beta+ np.sqrt(beta**2+ 8*alpha**2))
x=np.linspace(-10.0,10.0)
```

```python
plt.plot(x, 10.0*np.ones(len(x)), 'g-', label=r'$\lambda_0$')
plt.plot(x, y2(x,10.0), 'b:', label=r'$\lambda_1$')
plt.plot(x, y3(x,10.0), 'r-.', label=r'$\lambda_2$')
plt.xlabel(r'effective magnetic field $\alpha$', fontsize=13)
plt.ylabel(r'Energy levels')
plt.grid()
plt.title(r'$\beta=10$')
plt.legend(loc='best')
plt.savefig('energy_level_beta10.eps')
```

```python
plt.plot(x, 1.0*np.ones(len(x)), 'g-', label=r'$\lambda_0$')
plt.plot(x, y2(x,1.0), 'b:', label=r'$\lambda_1$')
plt.plot(x, y3(x,1.0), 'r-.', label=r'$\lambda_2$')
plt.xlabel(r'effective magnetic field $\alpha$', fontsize=13)
plt.ylabel(r'Energy levels')
plt.title(r'$\beta=1$')
plt.grid()
plt.legend(loc='best')
plt.savefig('energy_level_beta1.eps')
```
