```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def lambda_fn(t):
    p=np.pi
    lambda_0=0.1
    lambda_f=1.0
    return lambda_0 + (lambda_f-lambda_0)*np.sin(p/2*(np.sin(t*p/2)**2))**2
```

```python
t=np.linspace(0.1,1,100)
```

```python
plt.plot(t,lambda_fn(t), 'r')
```

```python
plt.plot(t, np.sin(np.sin(1./t)**2)**2)
```
