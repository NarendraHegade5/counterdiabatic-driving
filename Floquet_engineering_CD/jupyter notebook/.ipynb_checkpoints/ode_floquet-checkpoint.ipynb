```python
def SE_rhs(L, omega):
    """Return a function that evaluates the right-hand-side of the Schrod Eqn.
    (This can then be passed to an ODE solver.)
    Parameters
    ----------
    L : int
        Number of sites.
    Returns
    -------
    f : function
        A function that evaluates the RHS of the DNLS.
    """
    
    def f(t, y):
        u=np.exp(1j*(np.sin(omega*t))/omega)
        hoppings = u*np.diag(np.ones(shape=(L - 1,)),  1) + np.conj(u)*np.diag(np.ones(shape=(L - 1,)), -1)
        hoppings[0,L-1]=1.0 #periodic boundary condition
        hoppings[L-1,0]=1.0
        return  -1j*np.dot(hoppings, y)
    return f
```

```python
def psi_SE(tf,L=10,steps=1, omega=0, driving_amp=0):
    """
    Parameters
    ----------
    tf : float
        End time of the integration.
    L : int
        Number of lattice sites
    Returns
    -------
    y : float
    wavefn at time t
    """
    r = integrate.complex_ode(SE_rhs(L, omega)).set_integrator('dop853', nsteps=1000) #500 is default
    central_site_index = (L - 1)/2 #assuming odd number of sites
    ic = np.zeros(shape=(L,))#/np.sqrt(L)
    ic[central_site_index] = 1.0
    r.set_initial_value(ic)
    dt = float(tf)/steps
    wavefn = np.zeros((steps+1,L), dtype=np.complex_)
    t_arr=np.zeros((steps+1), dtype=np.float_)
    idx=0
    for idx in xrange(steps+1):
    #while r.t < tf:
        if r.successful(): 
                wavefn[idx,:] = r.y[:]
                t_arr[idx]=r.t
               #print idx
        else:
            raise ValueError("Integration failed at t = {}".format(r.t)) 
        r.integrate(r.t + dt)
        idx+=1
    return wavefn,t_arr
```

```python
from scipy import special
def magnus_Ham_0(drive_amp,omega,L):
    bessel_func=special.jv(0,drive_amp/omega)
    ham = bessel_func*(np.diag(np.ones(shape=(L - 1,)),  1) + np.diag(np.ones(shape=(L - 1,)), -1))
    ham[0,L-1]=1.0 #periodic boundary condition
    ham[L-1,0]=1.0
    return ham

def unitary_op(Ham, T):
    return sp.linalg.expm(-1j*Ham* T)

def psi_floquet(psi0,drive_amp, omega,L):
    Ham=magnus_Ham_0(drive_amp,omega,L)
    T=2*np.pi/omega
    U=unitary_op(Ham, T)
    return np.dot(U, psi0)

def fidelity(psi_1, psi_2):
    return np.absolute(np.vdot(psi_1, psi_2))**2
```
